We attempt to implement Quantum Dots Cellular Automata using IBM's Qiskit. We simulate a general gate which can be configured to simulate different types of gates such as X gate, Y Gate, Hadamard gate etc.

```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, plot_bloch_multivector
from qiskit.circuit.library import *
from qiskit.quantum_info import Statevector
import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.linalg import expm
from matplotlib.animation import FuncAnimation
# %matplotlib inline
```

```python
#Setting the constants

P0 = 3
E0 = 1
m = 4
t0 = 2*np.pi*m/E0

#constants for pi rotations
g_pi = np.pi/(2*t0)
P_pi = np.pi/(2*E0*t0)
```

```python
#Single gate matrices

#Pauli Gates
sigma_x = np.array([[0, 1],
                    [1, 0]])

sigma_y = np.array([[0, -1.j],
                    [1.j, 0]])

sigma_z = np.array([[1, 0],
                    [0, -1]])

#X Gates
def inv_xgate_g(g, t):
    theta = t*2*g
    gate = np.array([[np.cos(theta/2), np.sin(theta/2)*1.j],
                     [np.sin(theta/2)*1.j, np.cos(theta/2)]])
    return gate

def inv_xgate(theta):
    return inv_xgate_g(theta/(2*t0), t0)

#Z Gates
def zgate_P(P, t):
    theta = 2*E0*P*t
    gate = np.array([[np.exp(theta*1.j/2), 0],
                     [0, np.exp(-1*theta*1.j/2)]])
    return gate

def zgate(theta):
    return zgate_P(theta/(2*E0*t0), t0)

#Y Gates
def ygate(theta):
    z3pi_2 = zgate(1.5*np.pi)
    inv_xtheta = inv_xgate(theta)
    zpi_2 = zgate(np.pi/2)
    gate = zpi_2.dot(inv_xtheta).dot(z3pi_2)
    return gate

#general gate
def gen_gate(cvector, t):
    sum_sq = sum([ i*i for i in cvector ])
    sum_sq = np.sqrt(sum_sq)
    print('sum_sq', sum_sq)
    cvector = [i/sum_sq for i in cvector]
    sigma_n = cvector[0]*sigma_x + cvector[1]*sigma_y + cvector[2]*sigma_z
    theta = t*sum_sq
    return expm(-1.j*sigma_n*theta/2)

#Hadamard Gate
hgate = inv_xgate(np.pi).dot(ygate(np.pi/2))
```

```python
gen_gate([-2*g_pi, 0, 0], t0)
```

```python
gate = gen_gate([-1*g_pi, 0, 0], t0)

qc = QuantumCircuit(1)
# qc.h(0)
qc.unitary(gate, [0])

simulator = Aer.get_backend('statevector_simulator')
qresult = execute(qc, backend=simulator).result()
statevector = qresult.get_statevector()
plot_bloch_multivector(statevector)
```

```python
P1 = -3

theta_x = np.pi

g = theta_x/(4*np.pi*m)
```

```python
def E_sigma(E0, P0, P1, P2):
    return P0*E0 + P1*E0 + 2*P2*E0

rot_00 = gen_gate([-2*g, 0, -12*E0], t0)
rot_01 = gen_gate([-2*g, 0, -4*E0], t0)
rot_10 = gen_gate([-2*g, 0, -8*E0], t0)
rot_11 = gen_gate([-2*g, 0, 0], t0)*(-1.j)

rot_gate = np.zeros((8,8), dtype = complex)

rot_gate[0,0] = rot_00[0,0]
rot_gate[0,2] = rot_00[0,1]
rot_gate[2,0] = rot_00[1,0]
rot_gate[2,2] = rot_00[1,1]

rot_gate[1,1] = rot_01[0,0]
rot_gate[1,3] = rot_01[0,1]
rot_gate[3,1] = rot_01[1,0]
rot_gate[3,3] = rot_01[1,1]

rot_gate[4,4] = rot_10[0,0]
rot_gate[4,6] = rot_10[0,1]
rot_gate[6,4] = rot_10[1,0]
rot_gate[6,6] = rot_10[1,1]

rot_gate[5,5] = rot_11[0,0]
rot_gate[5,7] = rot_11[0,1]
rot_gate[7,5] = rot_11[1,0]
rot_gate[7,7] = rot_11[1,1]

rot_gate
```

```python
def unit_gate(mat):
    n = np.log2(len(mat))
    qc = QuantumCircuit(n)
    qc.unitary(mat, range(int(n)))
    gate = qc.to_gate()
    return gate

x_rot_pi = unit_gate(inv_xgate(np.pi))
x_rot_pi.name = 'NOT'

h_gate = unit_gate(hgate)
h_gate.name = 'H Gate'

d_gate = unit_gate(rot_gate)
d_gate.name = 'D Gate'
```

```python
qc = QuantumCircuit(3, 3)

qc.append(x_rot_pi, [2])
qc.barrier()

qc.i(0)
qc.append(x_rot_pi, [1])
qc.append(h_gate, [0])
qc.append(h_gate, [1])
qc.barrier()

# qc.cx(0,1)
qc.append(d_gate, [0,1,2])
qc.append(x_rot_pi, [1])
qc.barrier()

qc.append(h_gate, [0])
qc.measure([0], [0])

qc.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
qr = execute(qc, backend=simulator).result()
plot_bloch_multivector(qr.get_statevector())
```
